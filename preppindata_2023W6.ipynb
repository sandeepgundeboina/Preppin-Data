```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/drive/MyDrive/Abc/DSB Customer Survery.csv')
data
```

```python
mobile=data[['Customer ID','Mobile App - Ease of Use','Mobile App - Ease of Access','Mobile App - Navigation','Mobile App - Likelihood to Recommend']]
online=data[['Customer ID','Online Interface - Ease of Use','Online Interface - Ease of Access','Online Interface - Navigation','Online Interface - Likelihood to Recommend']]
```

```python
mobile=mobile.rename(columns={'Mobile App - Ease of Use':'Ease of Use','Mobile App - Ease of Access':'Ease of Access'
,'Mobile App - Navigation':'Navigation','Mobile App - Likelihood to Recommend':'Likelihood to Recommend'})
mobile['mb_Avg Rating']=mobile[['Ease of Use','Ease of Access','Navigation','Likelihood to Recommend']].mean(axis=1)
```

```python
online=online.rename(columns={'Online Interface - Ease of Use':'Ease of Use','Online Interface - Ease of Access':'Ease of Access',
             'Online Interface - Navigation':"Navigation",'Online Interface - Likelihood to Recommend':"Likelihood to Recommend"})
online['ol_Avg Rating']=online[['Ease of Use','Ease of Access','Navigation','Likelihood to Recommend']].mean(axis=1)
```

```python
mobile_f=mobile[['Customer ID','mb_Avg Rating']]
online_f=online[['Customer ID','ol_Avg Rating']]
```

```python
final_d=pd.merge(mobile_f,online_f,how='left',on='Customer ID')
final_d['diff']=final_d['ol_Avg Rating']-final_d['mb_Avg Rating']
```

```python
final_d['Preference']=final_d['diff'].apply(lambda x:'Mobile App SuperFan' if x<=-2 else('Mobile App Fan'if x<=-1  else
 ('Online Interface SuperFan' if x>=2 else ('Online Fan' if x >=1  else 'Neutral')) ))
```

```python
sol=final_d['Preference'].value_counts(normalize=True).reset_index(name='Percentage')
sol=sol.rename(columns={'index':'Preference'})
sol['Percentage']=(sol["Percentage"]*100).round(2)
```
