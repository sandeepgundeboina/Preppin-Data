```python
import pandas as pd
import numpy as np
import re
```

```python
d=pd.read_csv('/content/drive/MyDrive/Abc/preppin_w21/PD 2023 Wk 21 Input.csv')
d['student_id']=d['student_id'].astype('str')
```

```python
d_2021=d[['student_id', 'first_name', 'last_name', 'Gender', 'D.O.B',
       '2021-attainment', '2021-effort', '2021-attendance', '2021-behaviour']]
d_2022=d[['student_id', 'first_name', 'last_name', 'Gender', 'D.O.B',
          '2022-attainment', '2022-effort', '2022-attendance', '2022-behaviour']]
```

```python
d_2021['Overall_2021']=d_2021.mean(axis=1,numeric_only=True)
d_2022['Overall_2022']=d_2022.mean(axis=1,numeric_only=True)
```

```python
d_2021=d_2021[['student_id', 'first_name', 'last_name', 'Gender', 'D.O.B','Overall_2021']]
d_2022=d_2022[['student_id','Overall_2022']]
```

```python
gr=d_2021.merge(d_2022,how='inner',on='student_id')
gr.head(3)
```

```python
gr['diff']=gr['Overall_2022']-gr['Overall_2021']
gr['progress']=gr['diff'].apply(lambda x: 'Improvement' if x>0 else ('Cause of Concern' if x<0 else 'No Change'))
```

```python
grades=gr[gr['progress']=='Cause of Concern']
grades
```
