```python
import pandas as pd
import numpy as np
```

```python
all_df=pd.read_excel('/content/drive/MyDrive/Abc/preppin_w27/OverHeadCosts.xlsx',sheet_name=None)
df=pd.concat(all_df,ignore_index=True)
df=df.pivot_table(values=['Value'],columns=['Name'],index=['School Name','Year','Month']).reset_index(col_level=1)
df.columns=df.columns.droplevel()
```

```python
df['Day']='1'
df['Year']=df['Year'].astype('str')
df['Date']=df['Day']+'/'+df['Month']+'/'+df['Year']
df['Date']=pd.to_datetime(df['Date'])
```

```python
df=df[['School Name','Date','Electricity Cost', 'Gas Cost','Maintenance Cost', 'Water Cost' ]]
df['Total Cost']=df.sum(axis=1)
```

```python
f_df=df.sort_values(by=['School Name','Date']).reset_index().drop('index',axis=1)
f_df.head()
```
