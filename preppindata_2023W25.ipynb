```python
import pandas as pd
import numpy as np
import re
```

```python
east=pd.read_csv('/content/drive/MyDrive/Abc/preppin_w25/East Students.csv').drop('Unique Row ID',axis=1)
west=pd.read_csv('/content/drive/MyDrive/Abc/preppin_w25/West Students.csv').drop('Unique Row ID',axis=1)
school=pd.read_csv('/content/drive/MyDrive/Abc/preppin_w25/School Lookup.csv')
```

```python
east['Region']=east['Student ID'].apply(lambda x:x[:4])
east['Student ID']=east['Student ID'].apply(lambda x:x[4:])
east['Date of Birth']=pd.to_datetime(east['Date of Birth'])
east['FullName']=east['First Name']+' '+east['Last Name']
east=east[['FullName','Student ID','Date of Birth', 'Subject','Grade', 'Region']]
```

```python
west['Region']=west['Student ID'].apply(lambda x:x[-4:])
west['Student ID']=west['Student ID'].apply(lambda x:x[:5])
west['Date of Birth']=pd.to_datetime(west['Date of Birth'],format="%d/%m/%Y")
west['FullName']=(west['First Name']+' '+west['Last Name']).apply(lambda x:x.upper())
west['Grade']=west['Grade'].replace({1:'A',2:'B',3:'C',4:'D',5:'E',6:'F'})
west=west[['FullName','Student ID','Date of Birth', 'Subject','Grade', 'Region']]
```

```python
stud=pd.concat([east,west],ignore_index=True)
stud['Student ID']=stud['Student ID'].astype(str)
stud.tail(5)
```

```python
def f(a):
    if a=='A':
        return 50
    elif a=='B':
        return 40
    elif a=='C':
        return 30
    elif a=='D':
        return 20
    elif a=='E':
        return 10
    elif a=='F':
        return 0
```

```python
stud['Grade Score']=stud.Grade.apply(lambda x:f(x))
```

```python
df=stud.pivot_table(values=['Grade Score'],columns=['Subject'],index=['FullName','Student ID','Region','Date of Birth']).reset_index(col_level=1)
df.columns=df.columns.droplevel()
```

```python
df['Grade_score']=df[['English','Maths','Science']].mean(axis=1).round(0)
```

```python
def f(a):
    if a==50:
        return 'A'
    elif a==40:
        return 'B'
    elif a==30:
        return 'C'
    elif a==20:
        return 'D'
    elif a==10:
        return 'E'
    elif a==0:
        return 'F'
```

```python
for x in ['English','Maths','Science']:
    df[x]=df[x].apply(lambda k:f(k))
```

```python
school[['Student ID','School ID']]=school[['Student ID','School ID']].astype('str')
```

```python
admin=df.merge(school,how='inner',on='Student ID')
admin=admin[['Student ID','FullName','Date of Birth','Region','School Name','School ID','English', 'Maths','Science','Grade_score']]
```
