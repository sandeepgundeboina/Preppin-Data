```python
att=pd.read_excel('/content/drive/MyDrive/Abc/preppin_w22/Student Attendance vs Scores.xlsx',sheet_name='Attendance Figures')
score=pd.read_excel('/content/drive/MyDrive/Abc/preppin_w22/Student Attendance vs Scores.xlsx',sheet_name='Student Test Scores',header=1)
```

```python
att['attendance_percentage']=att['attendance_percentage']*100
```

```python
stud=att.merge(score,how='inner',on='student_name')
stud.head(3)
```

```python
stud['subject']=stud.subject.replace({'Math':'Maths','Engish':'English','Sciece':'Science'})
```

```python
stud['test_score_Integer']=stud['test_score'].round(0)
```

```python
stud['First Name']=stud['student_name'].str.split('_',expand=True).get(0)
stud['Surname']=stud['student_name'].str.split('_',expand=True).get(1)
```

```python
stud['Attendance Flag']=stud['attendance_percentage'].apply(lambda x: 'Low Attendance' if x<=70 else('High Attendance' if x>=90 else 'Medium'))
stud
```

```python
student=stud[['student_id','First Name','Surname','attendance_percentage','Attendance Flag','test_score','test_score_Integer']]
```
