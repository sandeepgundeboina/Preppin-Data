```python
import pandas as pd
import numpy as np
import re
```

```python
m_price=pd.read_excel('/content/drive/MyDrive/Abc/preppin_w20/Dining_Hall_Debacle_Input_Data.xlsx')
m_price.head(4)
```

```python
m_info=pd.read_excel('/content/drive/MyDrive/Abc/preppin_w20/Dining_Hall_Debacle_Input_Data.xlsx',sheet_name='Meal Nutritional Info')
m_info.head(3)
```

```python
meals=m_price.merge(m_info,how='left',on="Meal Option")
meals['Type']=meals['Type'].replace({'Veggie':'Vegetarian','Meat-based':'Meat based'})
```

```python
x=meals.groupby('Type').Price.agg(['sum','mean']).round(2).reset_index()
```

```python
x=x.rename(columns={'sum':'Total Cost','mean':'Avg price'})
```

```python
x['% of Total']=x['Total Cost'].apply(lambda y: (y/x["Total Cost"].sum())*100).round(2)
```

```python
final_meals=x[['Type','Avg price','% of Total']]
final_meals
```
