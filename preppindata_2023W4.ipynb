```python
import pandas as pd
import numpy as np
import calendar
```

```python
df={}
xls = pd.ExcelFile('/content/drive/MyDrive/Copy of New Customers.xlsx')
s2=[]
s2=pd.DataFrame(s2,columns=['ID','Joining Date','Account Type','Date of Birth','Ethnicity'])
for x in range(1,13):
   df['X_'+calendar.month_name[x]]=xls.parse(x-1)
   df['X_'+calendar.month_name[x]]['Month']=x
   df['X_'+calendar.month_name[x]]['Year']='2023'
   df['X_'+calendar.month_name[x]]['Joining Date']=df['X_'+calendar.month_name[x]]['Joining Day'].astype(str)+'/'+df['X_'+calendar.month_name[x]]['Month'].astype(str)+'/'+df['X_'+calendar.month_name[x]]['Year']
   df['X_'+calendar.month_name[x]]=df['X_'+calendar.month_name[x]].drop(['Joining Day','Month','Year'],axis=1)
   df['X_'+calendar.month_name[x]]['ID']=df['X_'+calendar.month_name[x]]['ID'].astype(str)
   df1=df['X_'+calendar.month_name[x]]
   s1=df1.pivot(index=['ID','Joining Date'], columns=['Demographic'], values='Value').reset_index().rename_axis(None, axis=1)
   s2=pd.concat([s2,s1],ignore_index=True)
```
