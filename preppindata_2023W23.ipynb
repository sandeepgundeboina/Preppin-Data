```python
stud=pd.read_excel('/content/drive/MyDrive/Abc/preppin_w23/Student Input.xlsx',sheet_name='Student Info')
res=pd.read_excel('/content/drive/MyDrive/Abc/preppin_w23/Student Input.xlsx',sheet_name='Results')
```

```python
stud[' Student ID ']=stud[' Student ID '].astype('str')
res['Student ID']=res['Student ID'].astype('str')
```

```python
res.head(4)
```

```python
stud_r=stud.merge(res,how='left',left_on=' Student ID ',right_on='Student ID')
stud_r=stud_r[[' Full name         ', ' Gender ', ' Class ',
       'Student ID', 'English', 'Economics', 'Psychology']]
```

```python
stud_r.head(5)
```

```python
d=stud_r.groupby(' Class ').mean()
```

```python
y=pd.DataFrame(d.idxmin(axis=0)).reset_index()
y.columns=['Subject',"Class"]
y
```

```python
x=pd.DataFrame(d[['English', 'Economics', 'Psychology']].min()).reset_index()
x.columns=['Subject','Grade']
x
```

```python
fdf=x.merge(y,how='left',on='Subject')
fdf
```
