```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/PD_2024_Wk_1_Input.csv')
data.head()
```

```python
new_columns = data['Flight Details'].str.split('//', expand=True)
new_columns.columns = ['Date', 'Flight_Code', 'From-To', 'Class', 'Price']
data = pd.concat([data, new_columns], axis=1)
data.head()
```

```python
flt_tofro=data['From-To'].str.split('-', expand=True)
flt_tofro.columns = ['From', 'To']
data = pd.concat([data, flt_tofro], axis=1)
data=data.drop(['Flight Details','From-To'],axis=1)
data.head()
```

```python
data['Date']=pd.to_datetime(data['Date'])
data['Price']=data['Price'].astype(float)
data.head()
```

```python
data['Flow Card?']=data['Flow Card?'].replace({1:'Yes',0:'No'})
data.head()
```
