```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df=pd.read_csv('PD 2022 Wk 1 Input - Input.csv')
df
```

```python
df['Pupils Name']=df['pupil last name']+","+df['pupil first name']
```

```python
df['Parental Contact First Name']=np.where(df['Parental Contact']==1,df['Parental Contact Name_1'],df['Parental Contact Name_2'])
```

```python
df['Parental Contact Full Name']=df['pupil last name']+","+df['Parental Contact First Name']
```

```python
df['Parental Contact Email address']=df['Parental Contact First Name']+'.'+df['pupil last name']+'@'+df['Preferred Contact Employer']
```

```python
df['Date of Birth']=pd.to_datetime(df['Date of Birth'])
```

```python
admitted=dt.datetime.strptime('08-30-2015','%m-%d-%Y')
```

```python
df['academic']=admitted-df['Date of Birth']
```

```python
df['academic']= df['academic'] / np.timedelta64(1, 'Y')
```

```python
df['Academic Year'] = df['academic'].apply(np.ceil).astype(int)
```

```python
df=df.drop(['id', 'pupil first name', 'pupil last name', 'gender', 'Date of Birth',
       'Parental Contact Name_1', 'Parental Contact Name_2',
       'Preferred Contact Employer', 'Parental Contact','Parental Contact First Name','academic'],axis=1)
```
