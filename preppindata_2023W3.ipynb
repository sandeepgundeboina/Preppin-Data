```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/Abc/PD 2023 Wk 1 Input.csv')
```

```python
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'])
```

```python
data['Bank']=data['Transaction Code'].str.split(pat='-',expand=True).get(0)
data['Online or In-Person']=data['Online or In-Person'].replace([1,2],['Online','In-Person'])
data
```

```python
df=data[data['Bank']=='DSB']
df['Quarter'] = df['Transaction Date'].dt.quarter
```

```python
sol=df.groupby(by=['Online or In-Person','Quarter'])['Value'].sum().reset_index(name='Value').rename(columns={'level_1':'Quarter'})
```

```python
target=pd.read_csv('/content/drive/MyDrive/Abc/Targets.csv')
target=target.set_index(['Online or In-Person']).stack().reset_index(name='Target').rename(columns={'level_1':'Quarter'})
target['Quarter']=target['Quarter'].replace(['Q1','Q2','Q3','Q4'],[1,2,3,4])
```

```python
tr1= sol.merge(target, on=['Online or In-Person','Quarter'])
```

```python
tr1['Varince to Traget']=tr1['Value']-tr1['Target']
tr1.sort_values(by="Varince to Traget",ascending=False)
```
