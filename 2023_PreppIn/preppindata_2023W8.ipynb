```python
import pandas as pd
import glob
import os
```

```python
path = r'/content/drive/MyDrive/Abc/preppin_w8/' # use your path
all_files = glob.glob(os.path.join(path , "*.csv"))

li = []

for filename in all_files:
    df = pd.read_csv(filename)
    try:
        if int(filename[-6:-4])<0:
            df['Month']=('1/'+filename[-5:-4]+'/2023')
        else:
            df['Month']=('1/'+filename[-6:-4]+'/2023')
    except:
        df['Month']='1/1/2023'
    li.append(df)
```

```python
f=[]
f=pd.DataFrame(f,columns=['id','first_name','last_name','Ticker	Sector','Market','Stock Name','Market Cap','Purchase Price','Month'])
for a in li:
    f=pd.concat([f,a],ignore_index=True)
```

```python
f=f.drop(['id','Ticker\tSector'],axis=1)
f=f.dropna()
f['Purchase Price']=f['Purchase Price'].str.split('$',expand=True) [1].astype('float')
f['Market Cap']=f['Market Cap'].str.split('$',expand=True)[1]
```

```python
f['Size']=f['Purchase Price'].apply(lambda x:'Low' if (x>0)and(x<25000) else('Medium' if (x>=25000)and(x<50000) else ('High' if (x>=50000)and(x<75000) else 'Very High')))
f=f.reset_index()
f=f.drop(['index'],axis=1)
```

```python
f['Market Cap']=f['Market Cap'].apply(lambda x:(float(x[:-1])*(10**9)) if x[-1]=='B' else (float(x[:-1])*(10**6)))
```

```python
f['Market Cap Size']=f['Market Cap'].apply(lambda x:'Small' if (x<(10**8))
                        else('Medium' if (x>=(10**8))and(x<(10**10)) else ('Large' if (x>=(10**10))and(x<(10**12)) else 'Huge')))
```

```python
f['Rank'] = f.groupby(['Month','Size','Market Cap Size'])['Purchase Price'].rank(ascending=False, method='dense').astype('int')
f=f[f["Rank"]<6].reset_index().drop('index',axis=1)
f.head(20)
```
