```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/Abc/PD 2023 Wk 1 Input.csv')
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'])
```

```python
data['Bank']=data['Transaction Code'].str.split(pat='-',expand=True).get(0)
data['Transaction Date'] = data['Transaction Date'].dt.month_name()
```

```python
data
```

```python
df=data[['Transaction Date','Bank','Value']]
df
```

```python
x=df.groupby(['Transaction Date','Bank']).sum('Value').reset_index()
```

```python
x['Rank'] = x.groupby(['Transaction Date'])['Value'].rank(ascending=False, method='dense')
y=x.groupby(['Bank'])['Rank'].mean().round(2).reset_index()
```

```python
x=x.merge(y,on='Bank',how='left',suffixes=('Rank', 'Avg_Rank'))
x=pd.DataFrame(x)
x=x.rename(columns={'RankRank':'Rank','RankAvg_Rank':'Avg_Rank'})
z=x.groupby('Rank')['Value'].mean().round(3).reset_index()
x=x.merge(z,on="Rank",how='left',suffixes=('Value', 'Avg_Value_per_Rank'))
x=pd.DataFrame(x)
Sol_fin=x.rename(columns={'ValueValue':'Value','ValueAvg_Value_per_Rank':'Avg_ValuePer_Rank'})
```
