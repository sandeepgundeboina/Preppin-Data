```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv('/content/drive/MyDrive/Abc/preppin_w15/Messy Nut House Data.csv')
```

```python
df['Location']=df['Location'].replace({'London':'London', 'L0ndon':'London', 'londen':'London', 'L0ndOn':'London', 'Manchester':'Manchester',
                                       'Manch3ster':'Manchester','MANchester':'Manchester', 'Livrepool':'Liverpool', 'Liverpool':'Liverpool',
                                       'LIverpool':'Liverpool', 'Liverp00l':'Liverpool'})
```

```python
fdf=pd.pivot_table(df,values=['Value'],index=['Location','Nut Type'],columns=['Category']).reset_index(col_level=1)
```

```python
fdf.columns=fdf.columns.droplevel()
```

```python
fdf['total cost']=fdf['Price (£) per pack']*fdf['Quant per Q']
fdf
```

```python
x=fdf.groupby('Location').agg({'Price (£) per pack':'mean',
                               'total cost':'sum'}).round(2).rename(columns={'Price (£) per pack':'Avg Price per Pack',
                                                                                                         'total cost':'Toatl Revenue'}).reset_index(col_level=1)
x.rename_axis(None, axis=1, inplace=True)
```
