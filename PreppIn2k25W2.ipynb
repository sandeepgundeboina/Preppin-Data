```python
import pandas as pd
import numpy as np
```

```python
d1=pd.read_csv('/content/drive/MyDrive/Abc/preppin24W2/PD 2024 Wk 1 Output Flow Card.csv')
d2=pd.read_csv('/content/drive/MyDrive/Abc/preppin24W2/PD 2024 Wk 1 Output Non-Flow Card.csv')
```

```python
df=pd.concat([d1,d2],axis=0)
df.head()
```

```python
df.dtypes
```

    Converting date column to datetime data type

```python
df['Date']=pd.to_datetime(df['Date'],dayfirst=True)
df.dtypes
```

    COnverting to Periods and getting quarter wise data based on year

```python
df['Quarter']=pd.PeriodDtype(freq='Q')
df['Quarter']=df['Date'].dt.to_period('Q')
df.head()
```

    Grouping data and pivoting and resetting index

```python
max=df.groupby(['Flow Card?','Quarter','Class'])['Price'].max().reset_index().pivot(index=['Flow Card?','Quarter'],columns='Class',values='Price').reset_index()
min=df.groupby(['Flow Card?','Quarter','Class'])['Price'].min().reset_index().pivot(index=['Flow Card?','Quarter'],columns='Class',values='Price').reset_index()
median=df.groupby(['Flow Card?','Quarter','Class'])['Price'].median().reset_index().pivot(index=['Flow Card?','Quarter'],columns='Class',values='Price').reset_index()
```

    Merging all data

```python
Dall=pd.concat([max,min,median],axis=0)
Dall.head()
```

    Changing column names and downloading as CSV

```python
Dall=Dall.rename(columns={'Business Class':'Premium','Economy':'First','First Class':'Economy','Premium Economy':'Business'})
Dall.to_csv('/content/sample_data/total_data.csv')
```
