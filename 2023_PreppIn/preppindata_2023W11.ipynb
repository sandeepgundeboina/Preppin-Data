```python
import numpy as np
import pandas as pd
from geopy import distance
```

```python
cus=pd.read_csv('/content/drive/MyDrive/Abc/preppin_w11/DSB Customer Locations.csv')
brch=pd.read_csv('/content/drive/MyDrive/Abc/preppin_w11/DSB Branches.csv')
```

```python
df=pd.DataFrame(columns=['Branch','Branch Lat','Branch Long','Distance','Customer','Address Lat','Address Long'])
for x in cus['Customer']:
    for y in range(0,4):
        a=(brch['Branch Lat'].iloc[y],brch['Branch Long'].iloc[y])
        b=(cus['Address Lat'].iloc[x-1],cus['Address Long'].iloc[x-1])
        new=pd.DataFrame({'Branch':[brch['Branch'].iloc[y]],'Branch Lat':[brch['Branch Lat'].iloc[y]],'Branch Long':[brch['Branch Long'].iloc[y]],
                          'Distance':[np.round((distance.distance(a,b).km),2)],'Customer':[cus['Customer'].iloc[x-1]],
                          'Address Lat':[cus['Address Lat'].iloc[x-1]],'Address Long':[cus['Address Long'].iloc[x-1]]})
        df=pd.concat([df,new], ignore_index=True)
```

```python
f=df.groupby(['Customer','Branch'])['Distance'].min().reset_index()
```

```python
f['Rank']=f.groupby(['Customer'])['Distance'].rank('dense')
f=f[f['Rank']==1.0].reset_index().drop(['index','Rank'],axis=1)
f['Priority']=f.groupby(['Branch'])['Distance'].rank('dense',ascending=False)
```

```python
final=pd.merge(f,df,how='left',left_on=['Customer','Branch','Distance'],right_on=['Customer','Branch','Distance'])
final=final[['Branch','Branch Lat','Branch Lat','Distance','Priority','Customer','Address Lat','Address Long']]
final.head(10)
```
