```python
import pandas as pd
import numpy as np
import warnings
```

```python
df=pd.read_csv('prepp2.csv')
df
```

```python
df['Pupils Name']=df['pupil first name']+" "+df['pupil last name']
```

```python
df['Date of Birth']=pd.to_datetime(df['Date of Birth'])
```

```python
df['Month']=df['Date of Birth'].dt.month_name()
```

```python
df['This years Birthday'] = df['Date of Birth'].apply(lambda x: x.replace(year = 2022))
```

```python
df['Cake needed on']=df['This years Birthday'].dt.day_name()
df['Cake needed on']=np.where(((df['Cake needed on']=='Saturday')|(df['Cake needed on']=='Sunday')),'Friday',df['Cake needed on'])
```

```python
d=df[['Pupils Name','Date of Birth','This years Birthday','Month','Cake needed on']]
```

```python
f={}
e=d[["Month",'Cake needed on']]
for x in e.Month.unique():
    f[x]=e[e['Month']==x]['Cake needed on'].value_counts()
```

```python
s=[]
for x,y in zip(d['Month'],d['Cake needed on']):
    s.append(f[x][y])
bd=pd.DataFrame(s)
d['BDs per weekday and month']=bd
```
